```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
df_X = df.drop(['Customer_Segment'], axis = 1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
corr_mat = df_X.corr()
```

```python
corr_mat
```

```python
eig_value, eig_vector = np.linalg.eig(corr_mat)
```

```python
eig_value
```

```python
eig_vector
```

```python
total = 0
var_sum = sum(eig_value)
var = []
for item in eig_value:
  L = (item/var_sum) * 100
  total = total + L

  if total < 75:
    var.append(L)
  else:
    break
```
